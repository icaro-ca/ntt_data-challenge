```python
from pyspark.sql.functions import current_timestamp
from pyspark.sql.types import (
    StructType,
    StructField,
    ArrayType,
    StringType,
    DateType,
    LongType,
    DoubleType
)
```

```python
CATALOG_NAME = 'covid_19'
SCHEMA_NAME = 'bronze'
TABLE_NAME = 'vaccinations'

FILE_PATH = '/Volumes/covid_19/bronze/raw_data/vaccinations.json'
```

```python
SCHEMA = StructType(
    [
        StructField('country', StringType(), True),
        StructField('data',
            ArrayType(
                StructType(
                    [
                        StructField('daily_people_vaccinated', LongType(), True),
                        StructField('daily_people_vaccinated_per_hundred', DoubleType(), True),
                        StructField('daily_vaccinations', LongType(), True),
                        StructField('daily_vaccinations_per_million', LongType(), True),
                        StructField('daily_vaccinations_raw', LongType(), True),
                        StructField('date', StringType(), True),
                        StructField('people_fully_vaccinated', LongType(), True),
                        StructField('people_fully_vaccinated_per_hundred', DoubleType(), True),
                        StructField('people_vaccinated', LongType(), True),
                        StructField('people_vaccinated_per_hundred', DoubleType(), True),
                        StructField('total_boosters', LongType(), True),
                        StructField('total_boosters_per_hundred', DoubleType(), True),
                        StructField('total_vaccinations', LongType(), True),
                        StructField('total_vaccinations_per_hundred', DoubleType(), True)
                    ]
                ),
            True),
        True),
        StructField('iso_code', StringType(), True)
    ]
)
```

```python
df = spark\
    .read\
    .format('json')\
    .schema(SCHEMA)\
    .option('multiLine', True)\
    .load(FILE_PATH)\
    .withColumn('ingestion_timestamp', current_timestamp())
```

```python
df.write\
    .mode('overwrite')\
    .saveAsTable(f'{CATALOG_NAME}.{SCHEMA_NAME}.{TABLE_NAME}')
```
