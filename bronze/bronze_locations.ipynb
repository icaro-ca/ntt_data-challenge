```python
from pyspark.sql.functions import current_timestamp
from pyspark.sql.types import StructType, StructField, StringType, DateType 
```

```python
CATALOG_NAME = 'covid_19'
SCHEMA_NAME = 'bronze'
TABLE_NAME = 'locations'

FILE_PATH = '/Volumes/covid_19/bronze/raw_data/locations.csv'
```

```python
SCHEMA = StructType(
    [
        StructField('location', StringType(), True),
        StructField('iso_code', StringType(), True),
        StructField('vaccines', StringType(), True),
        StructField('last_observation_date', DateType(), True),
        StructField('source_name', StringType(), True),
        StructField('source_website', StringType(), True)
    ]
)
```

```python
df = spark\
    .read\
    .format('csv')\
    .schema(SCHEMA)\
    .option('header', True)\
    .load(FILE_PATH)\
    .withColumn("ingestion_timestamp", current_timestamp())
```

```python
df.write\
    .mode('overwrite')\
    .saveAsTable(f'{CATALOG_NAME}.{SCHEMA_NAME}.{TABLE_NAME}')
```
