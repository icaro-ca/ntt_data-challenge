```python
from pyspark.sql import functions as F
```

```python
SOURCE_CATALOG_NAME = 'covid_19'
SOURCE_SCHEMA_NAME = 'silver'
SOURCE_TABLE_NAME = 'country_vaccine'

TARGET_CATALOG_NAME = 'covid_19'
TARGET_SCHEMA_NAME = 'gold'
TARGET_TABLE_NAME = 'dim_vaccine'
```

```python
df_dim_vaccine = (
    spark.table(f'{SOURCE_CATALOG_NAME}.{SOURCE_SCHEMA_NAME}.{SOURCE_TABLE_NAME}')
    .select('vaccine_name')
    .dropDuplicates()
    .withColumn(
        'vaccine_key',
        F.sha2(F.lower(F.trim('vaccine_name')), 256)
    )
    .select(
        'vaccine_key',
        'vaccine_name'
    )
)
```

```python
df_dim_vaccine\
    .write\
    .mode('overwrite')\
    .saveAsTable(f'{TARGET_CATALOG_NAME}.{TARGET_SCHEMA_NAME}.{TARGET_TABLE_NAME}')
```
