```python
from pyspark.sql import functions as F
```

```python
SOURCE_CATALOG_NAME = 'covid_19'
SOURCE_SCHEMA_NAME = 'silver'
SOURCE_TABLE_NAME = 'locations'

TARGET_CATALOG_NAME = 'covid_19'
TARGET_SCHEMA_NAME = 'silver'
TARGET_TABLE_NAME = 'country_vaccine'
```

```python
df = (
    spark.table(f'{SOURCE_CATALOG_NAME}.{SOURCE_SCHEMA_NAME}.{SOURCE_TABLE_NAME}')
    .select(
        'country',
        'iso_code',
        F.explode('vaccine_array').alias('vaccine_name')
    )
    .withColumn('vaccine_name', F.trim('vaccine_name'))
    .filter(F.col('vaccine_name').isNotNull())
    .filter(F.col('vaccine_name') != '')
    .dropDuplicates(['iso_code', 'vaccine_name'])
)
```

```python
df.write\
    .mode('overwrite')\
    .saveAsTable(f'{TARGET_CATALOG_NAME}.{TARGET_SCHEMA_NAME}.{TARGET_TABLE_NAME}')
```
