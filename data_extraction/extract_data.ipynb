```python
import requests

from enum import Enum
from dataclasses import dataclass
```

```python
dbutils.widgets.dropdown('file_type', 'locations', ['locations', 'vaccinations'])

file_type = dbutils.widgets.get('file_type')

if file_type not in ['locations', 'vaccinations']:
  raise ValueError(f'Invalid file type: {file_type}')
```

```python
VOLUME_PATH = '/Volumes/covid_19/bronze/raw_data/'
```

```python
@dataclass(frozen=True)
class DataParams:
    filename: str
    url: str

locations_params = DataParams(
    filename='locations.csv',
    url='https://raw.githubusercontent.com/owid/covid-19-data/refs/heads/master/public/data/vaccinations/locations.csv'
)
vaccinations_params = DataParams(
    filename='vaccinations.json',
    url='https://raw.githubusercontent.com/owid/covid-19-data/refs/heads/master/public/data/vaccinations/vaccinations.json'
)
```

```python
if file_type == 'locations':
  data_params = locations_params
elif file_type == 'vaccinations':
  data_params = vaccinations_params
```

```python
def get_data(url: str) -> bytes:
  response = requests.get(url)
  return response.content
```

```python
def save_data(data: bytes, file_path: str) -> None:
  with open(file_path, 'wb') as file:
    file.write(data)
```

```python

data = get_data(data_params.url)
save_data(data, VOLUME_PATH + data_params.filename)
```
